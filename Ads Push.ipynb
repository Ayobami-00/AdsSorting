```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import db
```

```python
### Fetch the service account key JSON file contents
cred = credentials.Certificate("serviceAccountKey.json")
```

```python
# Initialize the app with a service account, granting admin privileges
firebase_admin.initialize_app(cred, {
    'databaseURL': 'https://free-app-d2064.firebaseio.com/'
})
```

```python
##select users using version 2
selectedUsers = set()
ref = db.reference('users')
snapshot = ref.order_by_child('version').equal_to("version 2").get()
for key in snapshot:    
    print(key)
    selectedUsers.add(key)
```

```python
##select users that are Females.
selectedUsers2 = set()
ref = db.reference('users')
snapshot2 = ref.order_by_child('gender').equal_to("female").get()
for key in snapshot2:    
    print(key)
    selectedUsers2.add(key)
```

```python
selectedUsersAds = set.intersection(selectedUsers, selectedUsers2)
```

```python
def sendAds(selectedUsers, url, goto):
    for currentUser in selectedUsers:
        if currentUser:
            ref = db.reference('users')
            ref = db.reference('users').child(currentUser).child('ads').child('type').set("image")
            ref = db.reference('users').child(currentUser).child('ads').child('adsKey').set("05")
            ref = db.reference('users').child(currentUser).child('ads').child('gotoUrl').set(goto)            
            ref = db.reference('users').child(currentUser).child('ads').child('url').set(url)            
                
```
